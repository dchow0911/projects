            Introduction:

        </font>    
</h3>
<p>
    <span style='font-family:Georgia'>
    Default risk is the risk that a lender assumes in the event that a borrower won't be able to make the necessary payments on their debt obligation. Almost all forms of credit extensions subject lenders and investors to default risk. 
    The increasing trends in loan repayment default have been a matter of grave concern for policy makers and banks at large over a long period of time, defaulting on loan repayment not only affects safety of deposits, profitability of banks but the overall financial stability of a country as well. 
    This report aims to show how EDA can prove useful in the area of risk analytics in the banking and financial services sector by helping to minimize the risk of financial loss caused due to defaults on loan repayment. 
    The main purpose of the project is to understand how different parameters of the dataset under consideration are associated with loan defaults and derive conclusions based on the same thus helping in not only identifying the defaulting segments but also determining ways of minimizing losses due to loan repayment defaults.
    </span>
</p>

<a id="ub"></a>
<h3>
    <font color = purple >
        <span style='font-family:Georgia'>
            Understanding the Business:
            </span>   
        </font>    
</h3>
<p>
    <span style='font-family:Georgia'>
    Owing to the weak or non-existent credit histories, loan providers find it challenging to grant loans to individuals. Because of this, some customers take advantage of it by defaulting.
From a business perspective, banks are susceptible to broadly two types of risks in reference to providing loans to customers. These include:
    </span>
</p>
<ul>
    <span style='font-family:Georgia'>
        <li><b>Losing business due to loan refusal:</b> If the borrower is likely to repay the loan, refusing to grant it can result in the company losing business.</li>
        <li><b>Losing business due to loan grant:</b> If the borrower is not likely to pay back the loan, or is likely to default, then approving the loan may result in a loss of revenue for the business.</li>
    </span>
</ul>
    
<p><span style='font-family:Georgia'>In case of any loan application, there can be two scenarios of repayment, these are:
</span></p>
<ul>
    <span style='font-family:Georgia'> 
        <li><b>Payment difficulties--Late Payments:</b> When the loanee makes late payments of more than say m days on at least one of the first n installments of the loan for the sample under consideration.
</li>
        <li><b>Payment on Time:</b> When loanee makes timely installment payments.</li>
    </span>
</ul>
    
<p><span style='font-family:Georgia'>There are four possible decisions that a client or corporation can make in response to a loan application:
</span></p>

<ol>
    <span style='font-family:Georgia'>
        <li><b>Approved:</b> Loan Application accepted by Company</li>
        <li><b>Cancelled:</b> During approval, the client cancelled the application either due to a change of mind in regards to the loan, or certain circumstances posing greater risks or owing to the high loan prices.</li>
        <li><b>Refused:</b> Financing offer is refused as the client didn’t meet the necessary criteria.</li>
        <li><b>Unused offer:</b> The consumer cancels the loan at different phases of the procedure.</li>
    </span>
</ol>

```python
X = sm.add_constant( data['AMT_INCOME_TOTAL'] )
print("Feature Set")
print(X.head(10))
```

```python
print("Outcome Variable")
Y = data['AMT_ANNUITY']
print(Y.head(10))
```

```python
# Splitting the dataset into training and validation sets
train_X, test_X, train_y, test_y = train_test_split( X , Y, train_size = 0.8, random_state = 100 )
```

```python
# Datatypes

print("train_X")
print(train_X.info())
print("___________________________________________________________________")
print("test_X")
print(test_X.info())
print("___________________________________________________________________")
print("train_y",type(train_y))
print(len(train_y))
print("___________________________________________________________________")
print("test_y",type(test_y))
print(len(test_y))
```
